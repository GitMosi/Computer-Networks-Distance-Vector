```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a pandas DataFrame
filepath = r"C:\Users\mosi1\Python Projects\CE - Network 2\Distance Vector Project\Execution Times - Number of Nodes.xlsx"
df = pd.read_excel(filepath)

# Create a figure and axis object with specified figure size
fig, ax = plt.subplots(figsize=(10, 20))

# Plot the "Number of Nodes" against "Vector Distance Execution Time (seconds)"
ax.scatter(df['Number of Nodes'], df['Vector Distance Execution Time (seconds)'], color='r', label='Vector Distance')

# Overlay the "Number of Nodes" against "Dijkstra Execution Time (seconds)" on the same plot
ax.scatter(df['Number of Nodes'], df['Dijkstra Execution Time (seconds)'], color='b', label='Dijkstra')

# Set the labels for the x and y axes, and display the legend
ax.set_xlabel('Number of Nodes')
ax.set_ylabel('Execution Time (seconds)')
ax.legend()

# Adjust the y-axis limit
ax.set_ylim(bottom=0.00005, top=0.01) # adjust these numbers as needed

# Change the y-axis to logarithmic scale
ax.set_yscale('log')

plt.show()
```
